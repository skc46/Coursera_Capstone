#### Data Cleaning

Let's only keep the columns that we are intersted.

```python
columns_to_keep = ['ID', 'Severity', 'Start_Time', 'End_Time',
       'Start_Lat', 'Start_Lng', 'Distance(mi)',
       'Street', 'Side', 'City', 'County', 'State',
       'Zipcode', 'Country', 'Timezone',
       'Temperature(F)',
       'Visibility(mi)', 'Wind_Speed(mph)',
       'Precipitation(in)', 'Weather_Condition','Crossing',
       'Junction', 'Roundabout',
       'Traffic_Signal', 'Turning_Loop',
       'Sunrise_Sunset']

df1 = df[columns_to_keep]
df1.shape
```

```python
# to find the actual delay in terms of hours
df1['Start_Time'] = pd.to_datetime(df1['Start_Time'], format = "%Y/%m/%d %H:%M:%S")
df1['End_Time'] = pd.to_datetime(df1['End_Time'], format = "%Y/%m/%d %H:%M:%S")
df1['Time_Delay (Hrs)'] = (df1['End_Time'] - df1['Start_Time'])/np.timedelta64(1, 'h')
df1.head(10)
```

```python
# Let's get day of week
df1['Day_Of_Week'] = df1['Start_Time'].dt.weekday
#df1['Name_Of_Day'] = df1['Start_Time'].dt.weekday_name
df1['Day_Of_Week'].value_counts()
```

##### Dealing with missing values

```python
# Let's count the number of missing values in each columns
missing_data = df1.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

Since the __Precipitation__ column is missing 2,025,874 values, hence instead drop all rows we want to replace the missing values with mean.

```python
average = df1['Precipitation(in)'].astype('float').mean()
df1['Precipitation(in)'].replace(np.nan, average, inplace=True)
```

For missing values among other columns, we simply drop the rows containing the missing values.

```python
df1.dropna(subset = ['City', 'Temperature(F)', 'Visibility(mi)', 'Wind_Speed(mph)','Weather_Condition', 'Sunrise_Sunset'], axis = 0)
```
